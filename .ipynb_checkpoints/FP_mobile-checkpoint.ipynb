# Strength of Friendship Paradox - Mobile Phone Script
This notebook processes Copenhagen Network Study data from Sapiezynski et al. (2019). Particuarly, I analyze the mobile phone communication network here. 

```python
# import libraries
import pandas as pd
import numpy as np
from collections import Counter
from math import log10
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read data
df = pd.read_csv('../strength-friendship-paradox/data/cns/calls.csv')
```

```python
df
```

```python
# change -1 to 0 
# -1 indicates a missing call
df['duration'] = df['duration'].replace(-1,0)
```

```python
df[df.duplicated(subset=['caller','callee'], keep=False)]
```

```python
df_from = df.groupby(['caller','callee']).agg('sum').reset_index().rename(columns={'duration':'d_from'})
df_from
```

```python
df_to = df.groupby(['caller','callee']).agg('sum').reset_index().rename(columns={'caller':'callee',
                                                                                 'callee':'caller',
                                                                                 'duration':'d_to'})
df_to
```

```python
df_merge = pd.merge(df_from[['caller','callee','d_from']],
                    df_to[['caller','callee','d_to']],
                    on = ['caller','callee'], how = 'outer',
                    indicator=True)
df_merge        
```

```python
# drop reciprocated edge rows
df_merge = df_merge[~df_merge[['caller','callee']].apply(frozenset, axis=1).duplicated()]
```

```python
df_merge['w'] = df_merge[['d_from','d_to']].sum(axis=1)
df_merge
```

```python
df = df_merge
```

```python
# change column names for the following functions
df = df.rename(columns={'caller':'node1',
                        'callee':'node2'})
```

```python
def FP(df):
    '''calculate the strength of FP index'''
    fp = 1 + (df['std']/df['mean'])**2
    return(fp)

def summary_stat(percent = .5, df = df):
    '''
    calculate summary degree stat
    input: cutoff e.g., .5, .6, 1
    '''
    if isinstance(percent, int):
        cutoff = percent
        print('the cutoff is',cutoff)
    else:
        # create a cutoff value by percentile
        cutoff = df['w'].quantile(percent)
    # calculate Mean and SD
    def sum_stat(degree_df):
        sum_df = degree_df.agg({'degree':['mean','std','min','max']})
        sum_df = sum_df.T.reset_index(drop=True)
        return(sum_df)
    # separate cases
    if percent == 1:
        # degree calculation
        A_degree = pd.DataFrame(list(Counter(df.iloc[:,'node1'].append(df.loc[:,'node2'])).items()))
        A_degree.columns = ["node", "degree"]
        sum_df = sum_stat(A_degree)
        sum_df['type'] = 'All'
        sum_S['n'] = len(A_degree.index)
        sum_df['n%'] = len(A_degree.index)/len(pd.unique(df.iloc[:,0:2].values.ravel('K')))*100
    else:
        # separate Strong vs. Weak
        S_df = df.loc[df.w >= cutoff,:]
        W_df = df.loc[df.w < cutoff,:]
        # degree calculaion
        S_degree = pd.DataFrame(list(Counter(S_df.loc[:,'node1'].append(S_df.loc[:,'node2'])).items()))
        S_degree.columns = ["node", "degree"]
        W_degree = pd.DataFrame(list(Counter(W_df.loc[:,'node1'].append(W_df.loc[:,'node2'])).items()))
        W_degree.columns = ["node", "degree"]
        sum_S = sum_stat(S_degree)
        sum_S['type'] = 'Strong'
        sum_W = sum_stat(W_degree)
        sum_W['type'] = 'Weak'
        sum_S['n'] = len(S_degree.index)
        sum_W['n'] = len(W_degree.index)
        sum_S['n%'] = len(S_degree.index)/len(pd.unique(df.iloc[:,0:2].values.ravel('K')))*100
        sum_W['n%'] = len(W_degree.index)/len(pd.unique(df.iloc[:,0:2].values.ravel('K')))*100
        sum_df = sum_S.append(sum_W)
    # add extra info
    sum_df['percentile'] = percent
    sum_df['cutoff'] = cutoff
    # sum_df['country'] = d_name.split(sep='_')[0]
    return(sum_df)

def degree_cutoff(df, cutoff):
    '''
    this method removes nodes that have greater than a cutoff value
    input: df = edgelist dataframe; cutoff = integer like 150 or percentile like 0.5
    '''
    # calculate degree of each node
    A_degree = pd.DataFrame(list(Counter(df.loc[:,'node1'].append(df.loc[:,'node2'])).items()))
    A_degree.columns = ["node", "degree"]
    if isinstance(cutoff, int):
        print('the cutoff D is',cutoff)
    else:
        # create a cutoff value by percentile
        cutoff = A_degree['degree'].quantile(cutoff)
        print('the cutoff D is',cutoff)
    # make a cut
    A_list = A_degree.loc[A_degree.degree <= cutoff,]
    # select only rows that are part of A_list['node']
    df = df.loc[(df['node1'].isin(A_list['node'].to_list())) & (df['node2'].isin(A_list['node'].to_list())),]
    return(df)

def weight_cutoff(df, cutoff):
    '''
    this method removes nodes that have greater than a cutoff value
    input: df = edgelist dataframe; cutoff = integer like 150 or percentile like 0.5
    '''
    if isinstance(cutoff, int):
        print('the cutoff w is',cutoff)
    else:
        # create a cutoff value by percentile
        cutoff = df['w'].quantile(cutoff)
        print('the cutoff w is',cutoff)
    # make a cut
    D_list = pd.unique(df.loc[df.w > cutoff,['node1','node2']].values.ravel('K'))
    # select only rows that are NOT part of D_list
    df = df.loc[(~df['node1'].isin(D_list)) & (~df['node2'].isin(D_list)),]
    return(df)

def resource_cutoff(df, cutoff):
    '''
    this method removes nodes that have greater than a cutoff value
    input: df = edgelist dataframe; cutoff = integer like 50
    '''
    # create a long list of nodes and their weight 
    R_df = pd.concat([df[['node1','w']].rename(columns={'node1':'node'}),
                      df[['node2','w']].rename(columns={'node2':'node'})],
                      ignore_index=True)
    # now sum up the weights by node
    R_groupby = R_df.groupby('node')['w'].sum().reset_index()
    # elminate those nodes who are above the cutoff value
    R_list = R_groupby.loc[R_groupby.w <= cutoff,'node'].to_list()
    # select only rows that are part of R_list
    df = df.loc[(df['node1'].isin(R_list)) & (df['node2'].isin(R_list)),]
    return(df)
```

```python
# resource cutoff
stat_df = pd.concat([summary_stat(percent = x, df = df) for x in [.99,.95,.90,.85]]) 
stat_df['FP'] = FP(stat_df)
stat_df['N'] = len(A_degree.index)
stat_df['N%'] = stat_df.n/stat_df.N*100
stat_df.round(2)
```
